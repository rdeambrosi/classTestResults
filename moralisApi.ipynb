```python
import time
import pandas as pd
from moralis import evm_api  # Make sure this import works as per your setup

api_key = "eyJhbGciOiJIUzI1NiIsInR5cCI6IkpXVCJ9.eyJub25jZSI6ImEwNWI1OWU1LTQ3MGYtNGY2Ni1iZTc1LWFjOWEzZDZiMDJkYiIsIm9yZ0lkIjoiMzgyMTI0IiwidXNlcklkIjoiMzkyNjQwIiwidHlwZUlkIjoiYTYzY2U1YTQtNmM3Ny00ZmUzLTg4ODQtOWMxMWMzYjc2NTE4IiwidHlwZSI6IlBST0pFQ1QiLCJpYXQiOjE3MTAwODg2OTIsImV4cCI6NDg2NTg0ODY5Mn0.Rf-Oy5OnqpZof-Dgsu2PAb5WO8rnuuLn1i1hDjMpVJ8"

# List of wallet addresses
wallet_addresses = [
    "0xa719Bd304847dc79F1e868AB9344277979D82d5E",
    "0xf471cB4AbB22E416E44533ac5B2C9Cf3747aC9d2",
    # Add more wallet addresses as needed
]

all_data = []  # List to store data for all wallets

for address in wallet_addresses:
    params = {
        "chain": "eth",
        "address": address
    }

    # Fetch token balances for the current wallet address
    result = evm_api.token.get_wallet_token_balances(api_key=api_key, params=params)

    # Assuming 'result' contains the list of tokens and their info
    for item in result:
        item['wallet_address'] = address  # Add wallet address to each entry
        all_data.append(item)

    time.sleep(1)  # Wait for 1 second before making the next request to avoid hitting rate limits

# Convert the collected data into a pandas DataFrame
df = pd.DataFrame(all_data)

print(df)

```

```python
import pandas as pd

# Assuming 'df' is your DataFrame from the previous steps

# Convert balance to float type to avoid 'Integer out of range' error
df['balance'] = pd.to_numeric(df['balance'], errors='coerce', downcast='float')

# Adjust the balance for decimals and convert it into a more readable format
df['adjusted_balance'] = df.apply(lambda row: row['balance'] / (10 ** row['decimals']), axis=1)

# Filter out possible spam
df_filtered = df[df['possible_spam'] == False]

# Sort by wallet address, then by token symbol
df_sorted = df_filtered.sort_values(by=['wallet_address', 'symbol'])

# Display the sorted, filtered DataFrame with adjusted balances
print(df_sorted[['wallet_address', 'symbol', 'name', 'adjusted_balance']])
```
